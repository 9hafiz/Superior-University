```python
import matplotlib.pyplot as plt

movies = [
    ("Eternal Sunshine of the Spotless Mind", 20000000),
    ("Memento", 9000000),
    ("Requiem for a Dream", 4500000),
    ("Pirates of the Caribbean: On Stranger Tides", 379000000),
    ("Avengers: Age of Ultron", 365000000),
    ("Avengers: Endgame", 356000000),
    ("Incredibles 2", 200000000)
]


def add_movies():
    num_movies = int(input("How many movies do you want to add? "))
    for _ in range(num_movies):
        name = input("Enter the movie name: ")
        while True:
            try:
                budget = int(input("Enter the movie budget: "))
                if budget < 0:
                    print("Budget cannot be negative.")
                else:
                    movies.append((name, budget))
                    break
            except ValueError:
                print("Invalid input. Please enter a valid budget.")


def calculate_average_budget():
    return sum(budget for _, budget in movies) / len(movies)


def find_high_budget_movies():
    average_budget = calculate_average_budget()
    return [(name, budget, budget - average_budget) 
            for name, budget in movies 
            if budget > average_budget]


def display_high_budget_movies(high_budget_movies):
    print("\nMovies with a budget above the average:")
    movie_names = []
    budgets = []

    for name, budget, difference in high_budget_movies:
        print(f"{name} exceeds the average budget by ${difference:,}.")
        movie_names.append(name)
        budgets.append(budget)

    plt.figure(figsize=(10, 6))
    plt.barh(movie_names, budgets, color='lightcoral')
    plt.xlabel('Budget ($)')
    plt.title('Movies with High Budgets')
    plt.tight_layout()
    plt.show()


def main():
    add_more = input("Would you like to add more movies? (yes/no): ")
    if add_more.lower() == "yes":
        add_movies()

    average_budget = calculate_average_budget()
    print(f"\nThe average movie budget is: ${average_budget:,.2f}")

    high_budget_movies = find_high_budget_movies()
    print(f"\nNumber of movies with budget above average: {len(high_budget_movies)}")

    display_high_budget_movies(high_budget_movies)


if __name__ == "__main__":
    main()
```
